```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

```python
df = fetch_california_housing(as_frame=True).frame
```

```python
df.head()
```

```python
np.random.seed(42)

x = np.random.rand(50, 1) * 100

y = 3.5 * x + np.random.randn(50, 1) * 20
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
y_train_scaled = scalar.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = scalar.transform(y_test.reshape(-1, 1))
```

```python
model1=LinearRegression()
model1.fit(X_train_scaled,y_train_scaled)
model2=LinearRegression()
model2.fit(X_train,y_train)
```

```python
y_pred1 = scalar.inverse_transform(model1.predict(X_test_scaled))
y_pred2 = model2.predict(X_test)
```

```python
print(y_pred1)
print(y_pred2)
```

```python
s1=r2_score(y_test,y_pred1)
s2=r2_score(y_test,y_pred2)
print(s1)
print(s2)
```

```python
mse1= mean_squared_error(y_test,y_pred1)
mse2= mean_squared_error(y_test,y_pred2)
print(mse1)
print(mse2)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(X_test, y_pred1, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
